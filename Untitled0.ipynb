```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
```

```python
#loading data
fpath = '/content/drive/MyDrive/CodingDojo/04-DataViz/Week13/Data/Chicago-Crime_2001.csv'
df = pd.read_csv(fpath)
df.head()
```

```python
df.info()
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

##TOPIC;1

```python
# Topic 1) Comparing Police Districts
# Most crimes in 2022
crime_data_2001 = df[df['Date'].dt.year == 2001]
district_crime_counts_2001 = crime_data_2001['District'].value_counts()
most_crime_district_2001 = district_crime_counts_2001.idxmax()
least_crime_district_2001 = district_crime_counts_2001.idxmin()
print("District with most crimes in 2001:", most_crime_district_2001)
print("District with least crimes in 2001:", least_crime_district_2001)
```

##TOPIC:3

```python
# Topic 3) Comparing AM vs. PM Rush Hour
# Filter AM and PM rush hour crimes
am_rush_hour_crimes = df[(df['Date'].dt.hour >= 7) & (df['Date'].dt.hour < 10)]
pm_rush_hour_crimes = df[(df['Date'].dt.hour >= 16) & (df['Date'].dt.hour < 19)]
```

```python
# Top 5 common crimes during AM and PM rush hours
top_5_am_crimes = am_rush_hour_crimes['Primary Type'].value_counts().head(5)
top_5_pm_crimes = pm_rush_hour_crimes['Primary Type'].value_counts().head(5)

print("Top 5 common crimes during AM rush hour:\n", top_5_am_crimes)
print("Top 5 common crimes during PM rush hour:\n", top_5_pm_crimes)
```

##top 5 common crime during AM rush hour are THEFT,BATTERY,CRIMINAL DAMAGE,BULGARY,OTHER OFFFENSE.

## top 5 common crime during PM rush hour are THEFT,BATTERY,CRIMINAL DAMAGE,NARCOTICS,ASSAULT.

```python
# Motor Vehicle Thefts comparison
motor_vehicle_thefts_am = am_rush_hour_crimes[am_rush_hour_crimes['Primary Type'] == 'MOTOR VEHICLE THEFT'].shape[0]
motor_vehicle_thefts_pm = pm_rush_hour_crimes[pm_rush_hour_crimes['Primary Type'] == 'MOTOR VEHICLE THEFT'].shape[0]

if motor_vehicle_thefts_am > motor_vehicle_thefts_pm:
    print("Motor Vehicle Thefts are more common during AM rush hour.")
else:
    print("Motor Vehicle Thefts are more common during PM rush hour.")
```

##TOPIC:4

```python
# Topic 4) Comparing Months
# Monthly crime analysis
monthly_crime_counts = df.resample('M', on='Date').size()
most_crime_month = monthly_crime_counts.idxmax()
least_crime_month = monthly_crime_counts.idxmin()

print("Month with the most crime:", most_crime_month.strftime('%B'))
print("Month with the least crime:", least_crime_month.strftime('%B'))
```

```python
# Individual crimes comparison
individual_monthly_crimes = df.groupby(['Primary Type', pd.Grouper(key='Date', freq='M')]).size().unstack(fill_value=0)
ax = individual_monthly_crimes.plot();
fig = ax.get_figure()
fig.autofmt_xdate()
```
